```python
#import libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
from importlib import reload
```

```python
#importing the datasets 
df_transfer = pd.read_csv('transfers_cleaned.csv')
df_sold = pd.read_csv('transfers_out.csv')
df_totalSpent_avgFinish = pd.read_csv('total_spent_average_finish.csv')
df_position_total_signed = pd.read_csv('position_total_signed.csv')
```

```python
#Finding out the average spending for each teams
average_spend = df_transfer.groupby('club_name')['transfer_fee(millions)'].mean()
print(average_spend)
```

```python
#Plotting a scatter graph to find the correlation between Premier League team's spending and success
total_spent = df_totalSpent_avgFinish['total_spent']
average_finish = df_totalSpent_avgFinish['average_finish']

plt=reload(plt)
plt.scatter(total_spent,average_finish,color = 'blue', edgecolors='black')
plt.xlabel('Total Spent (Millions)')
plt.ylabel('Average Finish')
plt.title('Scatter plot of teams average finish and their total spending')

plt.show()
```

```python
#Plotting a bar chart to find out the most in-demand player position 
player_position = df_position_total_signed['position']
total_signed = df_position_total_signed['total_signed']


plt=reload(plt)
plt.barh(player_position,total_signed, color = 'green', edgecolor = 'black')
plt.xlabel('Total Signed')
plt.ylabel('Player Position')
plt.title('Graph of Total Signed by each position')

plt.show()
```

```python
#Plotting a bar chart to find out the total signed players signed for each position in teams that finish in the top 4
df = pd.read_csv('totalSignedTop4.csv')
position = df['position']
total_signed = df['total_signed']

plt=reload(plt)
plt.barh(position,total_signed, color = 'orange', edgecolor = 'black')
plt.xlabel('Total Signed')
plt.ylabel('Player Position')
plt.title('Graph of Total Signed by each position for top 4 teams')

plt.show()
```

```python
#Plotting a bar chart to find out the total signed players signed for each position in teams that does not finish in the top 4
df = pd.read_csv('non_top4_spending.csv')
position = df['position']
average_spent= df['average_spent']

plt=reload(plt)
plt.barh(position,average_spent, color = 'maroon', edgecolor = 'black')
plt.xlabel('Average fee')
plt.ylabel('Player Position')
plt.title('Graph of average fee for each position for non-top 4 teams')

plt.show()
```

```python
#Plotting a bar chart to find out the average spending fee for each position in-top 4 teams
df1 = pd.read_csv('average_top4_spent.csv')
position = df1['position']
average_spent= df1['average_spent']

plt=reload(plt)
plt.barh(position,average_spent, color = 'blue', edgecolor = 'black')
plt.xlabel('Average Fee')
plt.ylabel('Player Position')
plt.title('Graph of average fee for each position for top 4 teams')

plt.show()
```

```python
#Displaying the frequency of players age 
df = pd.read_csv('age_distribution.csv')
df
```

```python
#Displaying the above table in a graph format 
age = df['age']
frequency = df['frequency']

plt.figure().set_figwidth(15)
plt.bar(age,frequency, color = 'purple', edgecolor = 'black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Graph of players age and the frequency')
plt.show()
```
